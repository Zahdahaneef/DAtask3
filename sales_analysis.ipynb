**Total Revenue**

```python
df['Revenue'] = df['Quantity'] * df['Price'] # add new column name revenue
df
```

```python
total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue) # calculate the total revenue
```

**Best selling Product and Revenue**

```python
best_product = df.groupby("Product")["Revenue"].sum().idxmax()
best_revenue = df.groupby("Product")["Revenue"].sum().max()

print("Best Product:", best_product)
print("Revenue:", best_revenue)
```

**Total Quantity Sold**

```python
total_quantity = df["Quantity"].sum()
print("Total Quantity Sold:", total_quantity)
```
